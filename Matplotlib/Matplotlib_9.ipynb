# Plotting Live Data in Real Time

Here we will plot data that is continuously being generated in real time. This real time plot will be great for plotting data that is changing frequently that you want to monitor.

Example: Website that shows real time youtube sbuxcriber count.

```python
#  Uptil now we had plotted static data 

import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

x_vals = [0, 1, 2, 3, 4, 5]
y_vals = [0, 1, 3, 2, 3, 5]

plt.figure(figsize=((10,8)))
plt.plot(x_vals, y_vals)


plt.tight_layout()
plt.show()
```

```python
#  Uptil now we had plotted static data 

import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []
 

index = count() 
# count function from itertools will count one number at a time and each time we get the next values
```

```python
def animate(i):
  x_vals.append(next(index))
  y_vals.append(random.randint(0, 5)) # this will append the number between 0 and 5
  plt.plot(x_vals, y_vals)

# We want to run this function every second and plot this values that are getting appended to our list.
# We can do this with the help of func animation class from the matplotlib module.

ani = FuncAnimation(plt.gcf(), animate, interval = 1000)
# Here in the FuncAnimation function first we have input figure that we want to animate
# Then input function that we want to run that is animate
# Then interval how often we want to run this function it is in millisecond so for 1 second we put the values as 1000.

# Now matplotlib will run this animate function every second.

plt.figure(figsize=((10,8)))
plt.tight_layout()

# Each time you run this command it will plot different graph
# If it gives different color lines then it is basically plotting new line without clearing the previous one
# For that we will use plt.cla() command to get the output of same color line. 
# Use this command above the  plt.plot(x_vals, y_vals)
# Keep running this cell to see the different output of the graph gives.
```
