# **Filling Area on Line Plots**

```python
# Here we will again see the line plot
# We will doing it with fill argument it will make our graph looks professional

import pandas as pd
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/gdrive')
```

```python
df=pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Matplotlib Github/data5.csv')
```

```python
df.head()
```

```python
# Drawing a lineplot

# Asigning column values to variable
ages = df['Age']
dev_salaries = df['All_Devs']
py_salaries = df['Python']
js_salaries = df['JavaScript']

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label = 'All Devs')

plt.plot(ages, py_salaries, label='Python')
# Here we have plotted graph of developer salary and age 


plt.legend() # Order is important for legend function. Here it will plot dev_salaries graph first and then age graph on the same graph.

plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')

plt.tight_layout()
plt.show()

# This type of graph we have already seen.
```

```python
# Now let's see how we can add some fills to our plot

ages = df['Age']
dev_salaries = df['All_Devs']
py_salaries = df['Python']
js_salaries = df['JavaScript']

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label = 'All Devs')

plt.plot(ages, py_salaries, label='Python') 

# We can create fill by fill_between function
plt.fill_between(ages, py_salaries)



plt.legend() # Order is important for legend function. Here it will plot dev_salaries graph first and then age graph on the same graph.

plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')

plt.tight_layout()
plt.show()
```

```python
# We can see through the graph by making it transparent using alpha function

ages = df['Age']
dev_salaries = df['All_Devs']
py_salaries = df['Python']
js_salaries = df['JavaScript']

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label = 'All Devs')

plt.plot(ages, py_salaries, label='Python') 
# plt.plot(x-axis, y1 value, y2 value, label=label)
# Note that y2 value is zero by default


# We can create fill by fill_between function
plt.fill_between(ages, py_salaries, alpha=0.25)


plt.legend() 


plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')


plt.tight_layout()
plt.show()
```

```python
# In the above graph our fill is defaulting to zero
# Which means it is filling from Python plot down the way to Zero(X-axis)

# Here we can also use some condition to get some useful information from the graph

# For example let's say instead of filling it to zero fill it to the median of the overall salary

ages = df['Age']
dev_salaries = df['All_Devs']
py_salaries = df['Python']
js_salaries = df['JavaScript']

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label = 'All Devs')

plt.plot(ages, py_salaries, label='Python') 
overall_median = 57287 # calculated using pandas operation. This is the median salary


# We can create fill by fill_between function
plt.fill_between(ages, py_salaries, overall_median, alpha=0.25)


plt.legend() 


plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')


plt.tight_layout()
plt.show()

# You can see from the graph that where exactly our python plot crosses the median salary
# When python plot is below the overall median salary it fill-up to that point and when 
# Our plot is above the overall median salary it fills down to that point.
```

```python
# Here we see how can we make our plot better by adding some conditions 
# Let's say we want to fill the graph above median salary with different color and below with different color

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label = 'All Devs')

plt.plot(ages, py_salaries, label='Python') 


# We can create fill by fill_between function
plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate=True , alpha=0.25)
# Interpolate is use to fill the area in the corner. You can see the difference by putting false values see the output.


plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries <= overall_median), interpolate=True, color='red', alpha=0.25) # Comment this section and see the ouput


plt.legend() 


plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')


plt.tight_layout()
plt.show()

```

```python
# We can also fill the area between the curve by just passing 

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label = 'All Devs')

plt.plot(ages, py_salaries, label='Python') 


plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries > dev_salaries), interpolate=True , alpha=0.25, label='Above Avg')

plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries <= dev_salaries), interpolate=True, color='red', alpha=0.25, label='Below Avg') 


plt.legend() 


plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')


plt.tight_layout()
plt.show()
```
