# **Scatter Plot**
Scatter plots are great when you want to show the relationship between the two sets of values and see if they are correlated.

```python

import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn') # We change our style from 'fivethirtyeight' to seaborn

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y, s = 100) # s argument is the size of the dot. Visit documentation: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html 


plt.tight_layout()

plt.show()

# We can see from the plot that there is no correlated between the two values.
```

```python
# We can also assign different numbers to this different possibilities and those will give you different colors on the scatter plots 
# Below the colors list that color number could be suppose how much satisfying they are with a certain product or services
# Each of this values in colors will correspond to the data point of x and y


plt.style.use('seaborn') 

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, c=colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.75) # The link to different color map is bottom of the notebook

# The lighter one appear in the plot are closer to zero and darker ones are closer to 10.
```

```python
# We can also set a color bar for easy understanding of what that plot means.

plt.style.use('seaborn') 

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, c=colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Satisfacation')
```

# Now plot some real data from the csv files 

```python
# Importing csv files

import pandas as pd
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/gdrive')
```

```python
# The CSV files contains the data of likes and views on the youtube video

df=pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Matplotlib Github/data7.csv')

df
```

```python
# Taking the column values and asigning the values to it.

plt.style.use('seaborn') 

view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']


plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)
# plt.xscale('log')
# plt.yscale('log')


plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()

# Since our dataset contains only 200 videos so we have 200 dots.
```

```python
# There is one outlier which skew our plot
# To handle this we can use log method to plot our graph


plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)
plt.xscale('log')
plt.yscale('log')


plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()

# We can see that the outlier don't skew the data as compared to above graph
```

```python
# To make it more attractive we can use colormap

plt.scatter(view_count, likes, c=ratio, cmap = 'summer', edgecolor='black', linewidth=1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label('Likes/dislikes Ratio')

plt.xscale('log')
plt.yscale('log')


plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()


# So using a scatter plot like this is a great way to get the correlation for the values that you are plotting out.
```
