# Plotting Time Series Data

```python
# Importing pandas and matplotlib library

import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates # for changing date time format 

plt.style.use('seaborn')


dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y) # date will be on x axis 

plt.tight_layout()

plt.show()
```

```python
# We can this with the help of line by using linestyle

plt.plot_date(dates, y, linestyle='solid')

plt.tight_layout()

plt.show()

# We can also change the marker to none
```

```python
# For more readability of the dates value on the x axis
# we can use gcf() function that is Get Current Figure and autoformat function

plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.tight_layout()

plt.show()

# You can see that this dates are rotated and have a proper alignment.
```

```python
# We can also change the format of date and time using dates from matplotlib

plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b, %d %Y') # b is for month, d is for Day and Y is for Year. From python documentation

plt.gca().xaxis.set_major_formatter(date_format) # seting the x axis values

plt.tight_layout()

plt.show()

# You can see the date format in the x axis
# For more to understand date and time format here is a link to the python documentation
# https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
```
