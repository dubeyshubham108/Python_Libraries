# **Subplots**

 If you want to work with plots in more object oriented manner it's best to create plots by using subplots method.

```python
# Importing pandas and pyplot from matplotlib


import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn') # Using seaborn style
```

```python
# Asigning column values to variables

ages = df['Age']
dev_salaries = df['All_Devs']
py_salaries = df['Python']
js_salaries = df['JavaScript']

plt.plot(ages, py_salaries, label='Python')
plt.plot(ages, js_salaries, label='JavaScript')

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

We are using the plt object that we have imported that from py.plot object. We never created thid object ourselves.

We can do this by creating plt = plot(). Here we have created an new instance of a plot,

The way to do this is called Stateful.

**Figure**: It is a container that is holding our plot. We can have multiple plots in a single figure.

```python
# We can recreate the same plot that we have currently created by instantiating a figure and an axis.

ages = df['Age']
dev_salaries = df['All_Devs']
py_salaries = df['Python']
js_salaries = df['JavaScript']


fig1, ax = plt.subplots() # Here our axis is only set to a single axis or a single plot.

# Subplots create a figure and then specify a certain number of rows and columns of axis. 
# If we don't pass in our number of rows and columns then by defaults it is one by one.
# One row and one column which is simply one axis.


# Now changing plt to ax
ax.plot(ages, py_salaries, label='Python')
ax.plot(ages, js_salaries, label='JavaScript')

ax.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax.legend()

# For title, xlabel and ylabel we will change it to set_title, set_xlabel, set_ylabel
ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.tight_layout() # It is a method for pyplot object so it will remain the same.

plt.show()  # Similarly for show method.


# So now we can see that we got that same plot as above but here we are using this new axis.
```

If this is same then what is the purpose of using it?
Well, we have all the data in a single plot and now we want to broke that data into multipleplots.

Let say we want "Python and Javascript" data on one plot.
And "All Developers" data on a single plot.

```python
fig1, ax = plt.subplots()
print(ax)

# We can see that there is only one axes
```

```python
# For 2 rows and 2 column we have nested list of values.
fig1, ax = plt.subplots(nrows=2, ncols=2)
print(ax)

```

```python
# Let's say we want 2 rows and one column

fig1, ax = plt.subplots(nrows=2, ncols=1)
print(ax)

# We can see that the output is the list of 2 values
```

```python
# This two plot will still be on our same figure
# We can unpack this two axis in a single variable by calling ax1 and ax2.

fig1, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
print(ax1)
print(ax2)

# Now we can see that we are getting those subplots individually
# You can see more about it in this video. https://www.youtube.com/watch?v=C-gEQdGVXbk&t=782s&ab_channel=CoreySchafer
```

```python
#

ages = df['Age']
dev_salaries = df['All_Devs']
py_salaries = df['Python']
js_salaries = df['JavaScript']

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)
# sharex will only label the bottom ticks for two rows
# Similarly we have sharey

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()

fig1.savefig('fig1.png')
fig2.savefig('fig2.png')

# here we have 2 plots on a single figure.
```

```python
# If we wants this two plot to be on two different figures instead of the same figure
# We can do this by just creating another figure and an axis the same as we did it before.


fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()

fig1.savefig('fig1.png')
fig2.savefig('fig2.png')

# We can see that now we get two figures here.

# we can also save this figure 
fig1.savefig('fig1.png')
fig2.savefig('fig2.png')
```
