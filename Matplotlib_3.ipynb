# **Pie Charts**
Here we will discuss how to use the pie chart in Matplotlib.

```python
import matplotlib.pyplot as plt

print(plt.style.available)
```

```python
plt.style.use("fivethirtyeight") # We have choosen the style of Pie plot. You can experiment with it by inserting different values in it.


slices = [60, 40]  # List of values whose Pie chart we have to make. It's not necessary that values to add upto hundred.
labels = ['Sixty', 'Forty']


plt.pie(slices, labels=labels) 
 

plt.title("My Awesome Pie Chart") # Title of our Pie chart
plt.tight_layout()  # this function adjust the graph in the required space.
plt.show()

#Try inserting some values in a list and see the changes on Pie chart.
```

```python
# We can also have a separator on the edges of the colors.

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart") 
plt.tight_layout()  
plt.show()

# You can use matplotlib documentation for more variation. Type 'Matplotlib Wedge' on google
```

```python
# Change the colors of the slices

slices = [70, 80, 30, 20]
labels = ['Banana', 'Apple', 'Orange', 'Graphes']
colors = ['blue', 'red', 'yellow', 'green']

plt.pie(slices, labels = labels, colors = colors, wedgeprops = {'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# We can also use hex color values instead of writing the names of the color 
# Colors
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

```python
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

plt.pie(slices, labels = labels, colors = colors, wedgeprops = {'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

#This look more professional as above figure.
# Prefer this method of using color or use color from choosing the style plot example "fivethirtyeight"
```

```python
# Now see a dataset of the people who use a particular language
# This data is collected from stackoverflow website and it is a 2019 data.

import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")


# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

plt.pie(slices, labels = labels, wedgeprops = {'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# This does not look good if you have so many data
# If you have so many data then bar chart is recommended
```

```python
# If we take only five values then it will look lot better as compared to this one

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java',]

plt.pie(slices, labels = labels, wedgeprops = {'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# Now if we want to highlight a particular slice from the figure we can use explode 

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java',]


#explode = [] # list of floats show how much we want to emphasized that particular slice. explode[0] will give same result.
# Now suppose we want to explode python from the pie graph
# As python is on the fourth position on the in the list of labels so we can explode like this
explode = [0, 0, 0, 0.1, 0]


plt.pie(slices, labels = labels, explode=explode, wedgeprops = {'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# That value in explode 0.1 means that it is 10% of the radius 
# Experiment it by inserting different values of explode and see the changes.
```

```python
# We can also add shadow to our chart by passing an argument shadow=True

plt.pie(slices, labels = labels, explode=explode, shadow=True, wedgeprops = {'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# It look little 3D chart.
```

```python
# We can also change exactly how the chart is rotated by passing startangle argument.

plt.pie(slices, labels = labels, explode=explode, shadow=True, startangle=90, wedgeprops = {'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# If we want to display the percentage that each slice is taking up we can pass autopct arguement.


plt.pie(slices, labels = labels, explode=explode, shadow=True, startangle=90, autopct='%1.1f%%', wedgeprops = {'edgecolor':'black'})
# You can find more about it from the documentation of matplotlib.

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
