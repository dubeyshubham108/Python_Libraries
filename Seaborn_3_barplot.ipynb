# Seaborn Barplot
You can find more about the barplot on the official documentation of seaborn library by clicking [here](https://seaborn.pydata.org/generated/seaborn.barplot.html).

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_df = sns.load_dataset('tips')
```

##sns.barplot() function

sns.barplot() function belongs to seaborn data visualization library to draw barplot.

Syntax: sns.barplot(

                                     x=None,
                                     y=None,
                                     hue=None,
                                     data=None,
                                     order=None,
                                     hue_order=None,
                                     estimator=<function mean at 0x0000026F155D02F0>,
                                     ci=95,
                                     n_boot=1000,
                                     units=None,
                                     orient=None,
                                     color=None,
                                     palette=None,
                                     saturation=0.75,
                                     errcolor=’.26′,
                                     errwidth=None,
                                     capsize=None,
                                     dodge=True,
                                     ax=None,
                                     **kwargs,
                                    )

```python
# Now if we want to see the how much male had paid and how much female had paid
# Then we can do this by passing the hue argument

sns.barplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex)
```

```python
# To avoid writting the name of the dataframe we can use data function

sns.barplot(x='day', y='total_bill', hue = 'sex', data=tips_df)
```

```python
# If you want to change the order of our graph we can do it using the order function
# We create a list by the name day_list

day_list = ['Sun', 'Thur', 'Fri', 'Sat']

sns.barplot(x='day', y='total_bill', hue = 'sex', data=tips_df, order=day_list)

# You can see the order of x axis
```

```python
# In graph if we want to change the way that now it is representing Male first and then female 
# If we want to change that then we can do it
# We want to first show female and then male for that we have a function hue_order

hue_order = ['Female', 'Male']

sns.barplot(x='day', y='total_bill', hue = 'sex', data=tips_df, hue_order=hue_order)
```

```python
# Now suppose if we want to change the reading of y axis
# We can do that by using the parameter estimator

sns.barplot(x='day', y='total_bill', hue = 'sex', data=tips_df, hue_order=hue_order, estimator=np.mean)

# Nothing change because the mean is 25 
```

```python
# Let's do it with max value

sns.barplot(x='day', y='total_bill', hue = 'sex', data=tips_df, hue_order=hue_order, estimator=np.max)
plt.show()
```

```python
# We can also change the black color line in the bar graph by using the conference interval

sns.barplot(x='day', y='total_bill', hue = 'sex', data=tips_df, ci=20)

# ci is for conference inteval
# Similarly if we don't want black color line then we can do that by putting ci=0
```

```python
# We can also use n_boot with conference interval

sns.barplot(x='day', y='total_bill', hue = 'sex', data=tips_df, hue_order=hue_order, estimator=np.max, n_boot=2)
```

```python
# Now if we want our plot to be in horizontal direction then we can do that by using orient parameter

sns.barplot(x='day', y='total_bill', hue = 'sex', data=tips_df, orient='v')


# While using orient parameter one thing should be kept in mind that the x variable should not be categorical.
# If we put orient='h' then it will give horizotal plot
```

```python
# If you want to plot this graph horizontally then we can do that by interchanging the x and y variables

sns.barplot(y='day', x='total_bill', hue = 'sex', data=tips_df)
```

```python
sns.barplot(x='total_bill', y='size', hue='sex', data=tips_df,orient='h')
```

```python
# If we want to change teh color

sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, color='g')
```

```python
# We can also use the palette parameter

sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, palette='hot')
```

```python
# We can also change different palette

sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, palette='magma')
```

```python
# We can also use the saturation parameter

sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, saturation=0.3)
```

```python
# We can also change the area bar by passing errcolor

sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, errcolor='0.5')
```

```python
# To increase the width of the errcolor we will use errwidth parameter

sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, errwidth=12)
```

```python
# To increase the capsize 4

sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, capsize=1)
```

```python
# To express male and female in a single plot we can use the dodge operation

sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, dodge=False)
```

```python
# kwargs arguement 
# Keyword Arguments parameter
kwargs = {'alpha':0.9, 'linestyle':':', 'linewidth':5, 'edgecolor':'k'}
 
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df,**kwargs)
```

```python
# Pass Keyword  argument as parameter
sns.set() # for darkgrid background
sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, alpha =.9, linestyle = "-.", linewidth = 3,
           edgecolor = "g")
```

```python
# Axes parameter
sns.set()
ax = sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, alpha =.9, linestyle = "-.", linewidth = 3,
           edgecolor = "g")
 
ax.set(title = "Barplot of Tips DataFrame",
      xlabel = "Days",
      ylabel = "Total Bill")
```

```python
# sns.set – for background dark grid style
# plt.figure() – for figure size
# plt.title() – for barplot title
# plt.xlabel() – for x-axis label
# plt.ylabel() – for y-axis label
# plt.savefig() – for save figure
# plt.show() – for show image only

# Example of Seaborn Barplot
sns.set()
plt.figure(figsize = (16,9))
 
sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, alpha =1, linestyle = "-.", linewidth = 3,
           edgecolor = "k")
 
plt.title("Barplot of Days and Total Bill", fontsize = 20)
plt.xlabel("Days", fontsize = 15)
plt.ylabel("Total Bill", fontsize = 15)
 
plt.savefig("Barplot of Days and Total Bill")
plt.show()
```
