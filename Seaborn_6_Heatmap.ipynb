# Correlation in Heatmap.
Correlation is a statistical measure of the relationship between two variables (X, Y)

```python
# Import Libraries

import seaborn as sns # for data visualization
import pandas as pd # for data analysis
import numpy as np # for numeric calculation
import matplotlib.pyplot as plt # for data visualization
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Github/Seaborn/Who_is_responsible_for_global_warming.csv")
```

```python
df.corr()
```

```python
plt.figure(figsize=(16,9))

ax = sns.heatmap(df.corr(), annot=True, linewidth=3)

ax.tick_params(size=10, color='b', labelsize=10, labelcolor='b')

plt.title("Heatmap of 'Who is responsible for global warming'", fontsize=25)

plt.show()
```

```python
# Now to see heatmap from different dataset we can load it from sklearn

from sklearn.datasets import load_breast_cancer

cancer_dataset = load_breast_cancer()

cancer_dataset

# from sklearn when we import dataset it will in the dictionary format.
```

```python
# Now we want to convert this data into dataframe

cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']],
                         columns = np.append(cancer_dataset['feature_names'],['target']))

cancer_df
```

```python
plt.figure(figsize=(25,25))

ax = sns.heatmap(cancer_df.corr(), annot=True, linewidth=3)

ax.tick_params(size=10, color='b', labelsize=10, labelcolor='b')

plt.title("Heatmap of 'Breast Cancer Patients'", fontsize=25)

plt.show()
```
