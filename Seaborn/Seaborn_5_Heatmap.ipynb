# Heatmap

Heatmap is the graphical representation of 2D(two dimensional) data. 
Each data value is represents in a matrix and it has a special color.

```python
# Import Libraries

import seaborn as sns # for data visualization
import pandas as pd # for data analysis
import numpy as np # for numeric calculation
import matplotlib.pyplot as plt # for data visualization
```

```python
# create numpy 2D array

array_2d = np.linspace(1,5,12).reshape(4,3) 
 
print(array_2d) # print numpy array
```

```python
# create heatmap

sns.heatmap(array_2d) 
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Github/Seaborn/Who_is_responsible_for_global_warming.csv")
```

```python
df.head()
```

```python
# set country name as index and drop Country Code, Indicator Name and Indicator Code
# To get the datarame that contains only numeric values.
 
df  = df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'], axis=1).set_index('Country Name')
df.head()
```

```python
# Create heatmap
 
plt.figure(figsize=(16,9))
sns.heatmap(df)
```

Syntax: sns.heatmap(
  
                                       data,
                                       vmin=None,
                                       vmax=None,
                                       cmap=None,
                                       center=None,
                                       robust=False,
                                       annot=None,
                                       fmt=.2g,
                                       annot_kws=None,
                                       linewidths=0,
                                       linecolor=white,
                                       cbar=True,
                                       cbar_kws=None,
                                       cbar_ax=None,
                                       square=False,
                                       xticklabels=auto,
                                       yticklabels=auto,
                                       mask=None,
                                       ax=None,
                                       **kwargs,
                                     )

```python
# Now if we want to set the values of colorbar

plt.figure(figsize=(16,9))
sns.heatmap(df, vmin = 0, vmax = 21)
plt.show()
```

```python
# We can also change the color of the heatmap

plt.figure(figsize=(16,9))
sns.heatmap(df, cmap="coolwarm")
plt.show()
```

```python
# We can also show values on our color box by passing the annot arguement

plt.figure(figsize=(16,9))
sns.heatmap(df, cmap="coolwarm", annot=True)
plt.show()
```

```python
# fmp parameter - add text on heatmap cell
 
annot_arr = np.array([['a00','a01','a02'],
                      ['a10','a11','a12'],
                      ['a20','a21','a22'],
                      ['a30','a31','a32']], dtype = str)
 
sns.heatmap(array_2d, annot = annot_arr, fmt="s") # s -string, d - decimal
plt.show()
```

```python
# annot_kws parameter
 
plt.figure(figsize=(16,9))
 
annot_kws={'fontsize':10, 
           'fontstyle':'italic',  
           'color':"k",
           'alpha':0.6, 
           'rotation':"vertical",
           'verticalalignment':'center',
           'backgroundcolor':'w'}
 
sns.heatmap(df, cmap="coolwarm",annot = True, annot_kws= annot_kws)
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.heatmap(df, cmap="coolwarm",annot = True, annot_kws= annot_kws, linewidths=4)
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.heatmap(df, cmap="coolwarm",annot = True, annot_kws= annot_kws, linewidths=4, linecolor='k')
plt.show()
```

```python
# hide color bar with cbar parameter 
  
plt.figure(figsize=(16,9))
 
sns.heatmap(df, cbar = False)
```

```python
# change style and format of color bar with cbar_kws parameter
   
plt.figure(figsize=(14,14))
 
cbar_kws = {"orientation":"horizontal", 
            "shrink":1,
            'extend':'min', 
            'extendfrac':0.1, 
            "ticks":np.arange(0,22), 
            "drawedges":True,
           }

sns.heatmap(df, cbar_kws=cbar_kws)
```

```python
# change y-axis labels using yticklabels parameter
 
plt.figure(figsize=(16,9))
 
country_code = ['USA', 'GBR', 'IND', 'CHN', 'RUS', 'AUS', 'FRA', 'DEU', 'CAN', 'BRA', 'ARG', 'PAK', 'NPL', 'BGD', 'JPN']
 
sns.heatmap(df, yticklabels = country_code)
```

```python
plt.figure(figsize=(16, 9))

ax = sns.heatmap(df)

ax.set(title="Heatmap of 'Who is responsible for global warming'",
       xlabel="CO2 emissions (metric tons per capita) Per year",
       ylabel="Country Name")

sns.set(font_scale=5)
```
