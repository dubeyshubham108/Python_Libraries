# Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_df = sns.load_dataset("titanic")
titanic_df
```

Syntax: sns.scatterplot(
  
                                           x=None,
                                           y=None,
                                           hue=None,
                                           style=None,
                                           size=None,
                                           data=None,
                                           palette=None,
                                           hue_order=None,
                                           hue_norm=None,
                                           sizes=None,
                                           size_order=None,
                                           size_norm=None,
                                           markers=True,
                                           style_order=None,
                                           x_bins=None,
                                           y_bins=None,
                                           units=None,
                                           estimator=None,
                                           ci=95,
                                           n_boot=1000,
                                           alpha=’auto’,
                                           x_jitter=None,
                                           y_jitter=None,
                                           legend=’brief’,
                                           ax=None,
                                           **kwargs,
                                          )

```python
# Draw scatter plot between age and fare

sns.scatterplot(x="age", y="fare", data=titanic_df)
plt.show()
```

```python
sns.scatterplot(x="age", y="fare", data=titanic_df, hue="sex")
plt.show()
```

```python
# To dividi our scatter plot into different category like man, woman and child 
# We can pass the style arguement with 'who' column

sns.scatterplot(x="age", y="fare", data=titanic_df, hue="sex", style="who")
plt.show()
```

```python
# To change the size of the plot we can pass the size arguement.

plt.figure(figsize=(15,10))
sns.scatterplot(x="age", y="fare", data=titanic_df, hue="sex", style="who", size="who")
plt.show()
```

```python
# To increse the size of the points we can pass sizes

plt.figure(figsize=(16,9))
sns.scatterplot(x="age", y="fare", data=titanic_df, hue="sex", style="who", size="who", sizes=(100,400))
plt.show()
```

```python
# To see how many person whether it is a male or a female were alive after the crash
# For that we can pass the alive column in hue and 

plt.figure(figsize=(16,9))
sns.scatterplot(x="who", y="fare", data=titanic_df, hue="alive", style="alive", size="who", sizes=(100,200))
plt.show()
```

```python
# Now to make our data look good we can pass the palette option

plt.figure(figsize=(16,9))
sns.scatterplot(x="who", y="fare", data=titanic_df, hue="alive", style="alive", size="who", sizes=(100,200), palette='inferno_r',alpha=.7)
plt.show()
```
