# Sorting Data

```python
import pandas as pd
```

```python
people = {
    "first": ["Corey", 'Jane', 'John', 'Adam'], 
    "last": ["Schafer", 'Doe', 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com', 'A@email.com']
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
# Sort this by last name

df.sort_values(by='last')
```

```python
# If we want to sort this in descending order 

df.sort_values(by='last', ascending=False)
```

```python
# Now let's suppose we want to sort last name first and then sort by first name

df.sort_values(by='last', ascending=False)
```

```python
# Sort last name in desending order and first name in ascending order

df.sort_values(by=['last', 'first'], ascending=[False, True], inplace=True)
```

```python
df
```

```python
# Now we can sort them back by using sort index method

df.sort_index()
```

```python
# Now if we simply wants to see sorted last name and not the entire method

df['last'].sort_values()

# This are sorted in ascending order
```
